```python
import pandas as pd 
```

```python
df = pd.read_excel(r"C:\Users\educa\Desktop\Udemy\Data-Analysis-Toolkit\Customer Call List.xlsx")
df
```

```python
df=df.drop_duplicates()
df
```

```python
df2=df.drop(columns='Not_Useful_Column')
df2
```

```python
#strip remove useless character in data
#df2["Last_Name"]= df2["Last_Name"].str.lstrip('/')
#f2["Last_Name"]= df2["Last_Name"].str.lstrip('...')
df2["Last_Name"]= df2["Last_Name"].str.strip('')
df2["Last_Name"]= df2["Last_Name"].str.strip('123._/')
df2
```

```python
df2["Phone_Number"]=df2["Phone_Number"].apply(lambda x : str(x))
df2["Phone_Number"]= df2["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' +x[6:10])
df2
```

```python
df2["Phone_Number"]=df2["Phone_Number"].str.replace('nan--', '')
df2["Phone_Number"]=df2["Phone_Number"].str.replace('N/a--', '')
df2
```
