useful libraries for visiulization (pandas, numpy, matplotlib) 

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df= pd.read_csv(r"C:\Users\educa\Desktop\Udemy\Data-Analysis-Toolkit\Ice Cream Ratings.csv")
df
```

```python
df= df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-ticks')
```

```python
df.plot(kind='line',title='Ice cream Rating', xlabel='Daily Rating', ylabel='Scores')
```

```python
df.plot(kind='line',title='Ice cream Rating', xlabel='Daily Rating', ylabel='Scores', subplots= True)
```

```python
df.plot(kind='bar',stacked=True)
```

```python
df.plot.barh(stacked=True)
```

```python
df['Flavor Rating'].plot(kind='bar',stacked=True)
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating',s=300)
```

```python
df.plot.hist(bins=15) #by default bins is 10
```

```python
df.plot.box()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Overall Rating', figsize=(15,8))
```
